```python
from neuroindex import NeuroIndex
import numpy as np

VECTOR_DIM = 112

ni = NeuroIndex(path="./ni_store", dim=VECTOR_DIM, cache_size=1000)

def random_embedding(dim):
    return np.random.rand(dim).astype("float32")

docs = [
    "Deep learning uses neural networks",
    "AI models need large datasets",
    "Machine learning transforms healthcare"
]

for text in docs:
    vec = random_embedding(VECTOR_DIM)
    ni.add_document(text=text, vector=vec, metadata={"source": "demo"})

query_vec = random_embedding(VECTOR_DIM)
results = ni.search(query_vector=query_vec, k=3)

for r in results:
    print(f"{r.source} | similarity={r.similarity:.3f} | {r.text}")

ni.close()
```
