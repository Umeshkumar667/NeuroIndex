```python
# -------------------------------
# Cell 1: Imports & Setup
# -------------------------------
from core import NeuroIndex, SearchResult
import numpy as np
import random
import string
import os

# Vector dimension (must match your NeuroIndex config)
VECTOR_DIM = 112

# Helper to generate random embeddings
def random_embedding(dim=VECTOR_DIM):
    return np.random.rand(dim).astype("float32")

# Helper to generate random sample documents
def random_text(length=8):
    words = ["AI", "machine", "learning", "neural", "network", "vector", "embedding", "FAISS", "data", "graph"]
    return " ".join(random.choices(words, k=length))

# Path to persist NeuroIndex
NI_STORE_PATH = "./ni_store"
os.makedirs(NI_STORE_PATH, exist_ok=True)
```

```python
# -------------------------------
# Cell 2: Initialize NeuroIndex
# -------------------------------
ni = NeuroIndex(
    path=NI_STORE_PATH,  # folder will be created automatically
    dim=VECTOR_DIM,
    cache_size=1000       # optional
)
print("âœ… NeuroIndex initialized")
```

```python
# -------------------------------
# Cell 3: Add 100 sample documents
# -------------------------------
NUM_DOCS = 100
doc_ids = []

for i in range(NUM_DOCS):
    text = random_text(length=10)
    vec = random_embedding()
    doc_id = ni.add_document(
        text=text,
        vector=vec,
        metadata={"source": "demo_doc", "doc_index": i}
    )
    doc_ids.append(doc_id)
    
print(f"âœ… Added {NUM_DOCS} documents to NeuroIndex")
```

```python
# -------------------------------
# Cell 4: Perform a sample query
# -------------------------------
# Example query
query_text = "neural network embedding"
query_vec = random_embedding()  # In real use, generate vector from your model

# Search top 5 results
results = ni.search(
    query_vector=query_vec,
    k=5
)

print("ðŸ” Search Results:")
for r in results:
    print(f"{r.source:6} | similarity={r.similarity:.3f} | {r.text} | metadata={r.metadata}")
```

```python
# -------------------------------
# Cell 5: Verify persistence
# -------------------------------
# Close & re-import NeuroIndex

ni2 = NeuroIndex(
    path=NI_STORE_PATH,
    dim=VECTOR_DIM
)

# Perform search again
query_vec2 = random_embedding()
results2 = ni2.search(
    query_vector=query_vec2,
    k=5
)

print("ðŸ”„ Persistence Test Results:")
for r in results2:
    print(f"{r.source:6} | similarity={r.similarity:.3f} | {r.text} | metadata={r.metadata}")
```

```python
# -------------------------------
# Cell 6: Inspect stats
# -------------------------------
stats = ni.stats()
print("ðŸ“Š NeuroIndex Stats:")
for key, value in stats.items():
    print(f"{key:15}: {value}")
```
